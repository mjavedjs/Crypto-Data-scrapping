```python
from bs4 import BeautifulSoup
import requests 
url = "https://coinmarketcap.com/" 
response = requests.get(url) 
print(response.status_code)
soup = BeautifulSoup(response.content, "html.parser")  #urllib 
```

```python
print(soup.prettify())
```

```python
table = soup.find('table')

if table:
    print("Table found! Getting all data...\n")
    
    # Get all rows
    rows = table.find_all('tr')
    
    # Process first 10 rows (skip header if exists)
    for row_index, row in enumerate(rows[1:100], 1):  # Skip header, get first 10
        # Get all td cells in this row
        cells = row.find_all('td')
        
        if cells:  # If row has cells
            print(f"=== Row {row_index} ===")
            
            # Extract data from each cell based on position
            for cell_index, cell in enumerate(cells):
                # Get all text from the cell, cleaning it up
                cell_text = cell.get_text(strip=True, separator=' ')
                
                # Remove extra whitespace
                cell_text = ' '.join(cell_text.split())
                
                if cell_text:  # Only print if there's content
                    # Label based on column position (you can customize these labels)
                    labels = ["#", "Name", "Price", "1h %", "24h %", "7d %", 
                             "Market Cap", "Volume(24h)", "Circulating Supply", "Last 7 Days"]
                    
                    if cell_index < len(labels):
                        print(f"{labels[cell_index]}: {cell_text}")
                    else:
                        print(f"Column {cell_index}: {cell_text}")
            
            print("-" * 50)
else:
    print("Table not found")
    # Try to find table by class if direct find doesn't work
    tables = soup.find_all('table')
    print(f"Found {len(tables)} tables on the page")
```
